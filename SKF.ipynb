```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import dates
import datetime
```

```python
df = pd.read_csv('ts_ax.csv')
```

```python
df
```

```python
#First method

#Advantage of this method is the most clear representation of the mixing in mm by time series.
#You can visually trace the anomalies and the dynamics of behavior.

origin = df['timestamp ']

a = [datetime.datetime.strptime(d, '%d/%m/%Y %H:%M') for d in origin]
  
b = df['Motor Axial vibration']

x = matplotlib.dates.date2num(a)
formatter = matplotlib.dates.DateFormatter('%d/%m')

figure = plt.figure(figsize=(10,10))
axes = figure.add_subplot(1, 1, 1)
  
axes.xaxis.set_major_formatter(formatter)
plt.setp(axes.get_xticklabels(), rotation = 15)
  
axes.plot(x, b)
plt.show()
```
